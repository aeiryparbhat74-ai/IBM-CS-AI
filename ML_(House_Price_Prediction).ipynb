```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("house_prices.csv")
```

```python
print(df.head())
```

```python
df.hist()
plt.suptitle("Histogram of Features")
plt.show()
```

```python
avg_price_bedrooms = df.groupby("Bedrooms")["Price"].mean()
avg_price_bedrooms.plot(kind='bar')
plt.title("Average Price by Bedrooms")
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.show()
```

```python
floor_counts = df["Floors"].value_counts()
plt.pie(floor_counts, labels=floor_counts.index, autopct='%1.1f%%')
plt.title("Distribution of Floors")
plt.show()
```

```python
corr = df.corr()
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python
X = df.drop("Price", axis=1)
y = df["Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
score = model.score(X_test, y_test)
print("Model Accuracy:", score)
```

```python
sample_house = [[1500, 3, 3, 2, 5, 8]]

predicted_price = model.predict(sample_house)
print("Predicted Price:", predicted_price[0])
```
